데이터 분할

```python
import os

base_path = 'C:/Users/konyang/dataset_yolo'

train_img_path = os.path.join(base_path, 'train/images')
train_lbl_path = os.path.join(base_path, 'train/labels')
val_img_path = os.path.join(base_path, 'val/images')

# Val에 있는 402장(시험 문제) 이름 리스트업
val_imgs = os.listdir(val_img_path)

deleted_count = 0
print("🧹 Train 폴더에서 중복된 Val 데이터(시험 문제)를 지우는 중...\n")

for img_name in val_imgs:
    # 1. Train 이미지 폴더에서 시험문제 삭제
    t_img = os.path.join(train_img_path, img_name)
    if os.path.exists(t_img):
        os.remove(t_img)
        deleted_count += 1

    # 2. Train 라벨 폴더에서 짝꿍(.txt) 삭제
    lbl_name = os.path.splitext(img_name)[0] + '.txt'
    t_lbl = os.path.join(train_lbl_path, lbl_name)
    if os.path.exists(t_lbl):
        os.remove(t_lbl)

print(f"✅ 정답 유출 방지 완료! 중복 데이터 {deleted_count}장 삭제됨.")
print("-" * 30)
print(f"🎯 [최종] Train 이미지: {len(os.listdir(train_img_path))}장 (정상)")
print(f"🎯 [최종] Val 이미지: {len(os.listdir(val_img_path))}장 (정상)")
```

```python
import os

base_path = 'C:/Users/konyang/dataset_yolo'

train_img_path = os.path.join(base_path, 'train/images')
val_img_path = os.path.join(base_path, 'val/images')

def count_images(folder_path):
    if not os.path.exists(folder_path):
        return 0
    return len([f for f in os.listdir(folder_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))])

train_cnt = count_images(train_img_path)
val_cnt = count_images(val_img_path)

print(f"✅ Train 이미지: {train_cnt}장")
print(f"✅ Val 이미지: {val_cnt}장")
print(f"🎯 총합: {train_cnt + val_cnt}장")
```
